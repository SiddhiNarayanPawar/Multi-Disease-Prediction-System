```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

```python
Dataset = pd.read_csv("Diabeties_Data.csv")
```

```python
#First 5 datset
Dataset.head()
```

```python
#Last 5 datset
Dataset.tail()
```

```python
#Number of Rows and Columns
Dataset.shape
```

```python
#GETTING THE STATISTICAL MESURES  FOR DATA
Dataset.describe()
```

```python
Dataset.describe()
```

```python
Dataset['Outcome'].value_counts()
```

```python
# 0 --- Diabetic 
Dataset.groupby('Outcome').mean()
```

```python
#Separating Data Labels and Variables
X = Dataset.drop(columns ='Outcome', axis=1)
Y = Dataset['Outcome']
```

```python
print(X)
```

```python
#Data Standarization
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X)
```

```python
print(scaler)
```

```python
standardized_data = scaler.transform(X)
```

```python
print(standardized_data)
```

```python
X = standardized_data
Y = Dataset['Outcome']
```

```python
print(X)
```

```python
print(Y)
```

```python
#Train-Test Split Data
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
classifier = svm.SVC(kernel = "linear")
```

```python
classifier.fit(X_train,Y_train)
```

```python
#Module Evaluation
```

```python
#accuaracy score on training data
X_train_prediction = classifier.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy Score of the training Data is ",training_data_accuracy)
```

Accuracy Score on Test_data


```python
X_test_prediction = classifier.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print("Accuracy Score of the testing Data is ",testing_data_accuracy)
```
